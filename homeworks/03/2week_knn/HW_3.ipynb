##### Реализуйте KNN. Сами, как делали это на лекции. 

Можете улучшить его.
(например использовать kd-tree для быстрого поиска ближайших соседей, 
 или обходить соседей и класть расстояния в мин-кучу - тогда вы быстрее будете набирать их, без сортировки.)

```python
class KNN:
    
    def __init__(self, n):
        self.n = n
        
    def fit(self, X, y):
        self.X = X
        self.y = y
    
    def predict(self, X):
        y_hat = []
        assert len(X.shape) == 2 #подается именно матрица
        print(len(self.X.shape))
        print(len(self.y.shape))
        #расстояние до всех в трен.выборке:
        for (h, w, u) in X:
            d=np.sqrt((h-self.X[:,0])**2+(w-self.X[:,1])**2+(u-self.X[:,2])**2)
            #индексы близжайших соседей
            idx = np.argsort(d)[:self.n]
            counts=np.bincount(self.y[idx])
            prediction = np.argmax(counts)
            y_hat.append(prediction)
        return np.array(y_hat)
```

```python
knn = KNN(5)

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
class KNN2:

    
    def __init__(self, n):
        self.n = n
        
    def fit(self, X, y):
        self.X = X
        self.y = y
    
    def predict(self, X):
        y_hat = []
        assert len(X.shape) == 2 #подается именно матрица
        #print(len(self.X.shape))
        #print(len(self.y.shape))
        #расстояние до всех в трен.выборке:
        for h in X:#векторы содержат уже не 3 компоненты, а больше
            d = 0
            for i in range(len(h)):
                d += (h[i] - self.X[:,i])**2
            d = np.sqrt(d)
            #индексы близжайших соседей
            idx = np.argsort(d)[:self.n]
            counts=np.bincount(self.y[idx])
            prediction = np.argmax(counts)
            y_hat.append(prediction)
        return np.array(y_hat)
```
